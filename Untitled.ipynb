**WE ARE ONLY ANALYSING THE INDIAN RESURANT SO WE USE COUNTRY CODE 1**

```python
Indian_res=df[df['Country Code']==1]
```

```python
Indian_res.head()
```

```python
city_votes=Indian_res.groupby(['City'])[['Votes']].sum()
```

```python
city_votes.reset_index(inplace=True)
```

```python
city_votes.sort_values(['Votes'],inplace=True)
```

```python
city_votes.head()
```

```python
plt.figure(figsize=(16,8))
sns.barplot(data=city_votes,x='City',y='Votes',log=True,  edgecolor='Red')
plt.xticks(rotation=90)
plt.xlabel('CITY',fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Brown'})
plt.ylabel('VOTES',fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Brown'})
plt.title('TOTAL VOTES ACROSS EACH CITY',fontdict={'fontsize': 20, 'fontweight' : 10, 'color' : 'Green'})
plt.show()
```

**1)MOHALI HAS LEAST NUMBER OF VOTES**


**2)NEW DELHI HAS HIGHEST NUMBER OF VOTES**

```python
Exce_rating=Indian_res[Indian_res['Rating text']=='Excellent']
Exce_rating
```

```python
plt.figure(figsize=(16,8))
sns.countplot(data=Exce_rating,x='City',hue='Rating text',log=True,edgecolor='Red',saturation=2,color='Yellow')
plt.xticks(rotation=90)
plt.title('Number Of  Excellent Rating Resturant In Each City',fontdict={'fontsize': 20, 'fontweight' : 10, 'color' : 'Green'})
plt.xlabel('CITY',fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Brown'})
plt.ylabel('COUNT',fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Brown'})
plt.show()
```

```python
Resuratnat_high_no_votes=Indian_res.groupby(['Restaurant Name'])[['Votes']].sum().reset_index().sort_values(['Votes'],ascending=False)
Top_ten_resturant_according_to_votes=Resuratnat_high_no_votes.head(10)
Top_ten_resturant_according_to_votes
```

```python
Resuratnat_high_no_votes=Indian_res.groupby(['Restaurant Name','City'])[['Votes']].sum().reset_index().sort_values(['Votes'],ascending=False)
Top_ten_resturant_according_to_votes_and_city=Resuratnat_high_no_votes.head(10).sort_values(['City'],ascending=False)
Top_ten_resturant_according_to_votes_and_city
```

```python
from collections import Counter
sd=Indian_res['Cuisines']
language_counter=Counter()
for i in sd:
    language_counter.update(i.split(','))
language=[]
popularity=[]
for item in language_counter.most_common(15):
    language.append(item[0])
    popularity.append(item[1])
```

```python
print(language)
print(popularity)
```
